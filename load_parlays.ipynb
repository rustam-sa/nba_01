```python
import pandas as pd
import glob
import os

# Path to the directory containing the CSV files (use raw string or double backslashes)
directory_path = r'E:\coding_projects\nba_01\parlays\5-16-2024'

# Use glob to get a list of all CSV files in the directory
csv_files = glob.glob(os.path.join(directory_path, '*.csv'))

# Debugging output to check if files are found
print(f"Found {len(csv_files)} CSV files.")

# List to hold DataFrames
dataframes = []

# Loop through the list of CSV files and read each one into a DataFrame
for file in csv_files:
    df = pd.read_csv(file, index_col=0)
    dataframes.append(df)

# Check if any dataframes were added
if dataframes:
    # Concatenate all DataFrames into a single DataFrame
    combined_df = pd.concat(dataframes, ignore_index=True)

    # Remove duplicate rows
    combined_df.drop_duplicates(inplace=True)

    # Optionally, reset the index if you want
    combined_df.reset_index(drop=True, inplace=True)

    # Now combined_df contains all the data from the CSV files with duplicates removed
    print(combined_df)
else:
    print("No dataframes to concatenate.")
```

```python
combined_df.to_csv("combined_df.csv")
```
