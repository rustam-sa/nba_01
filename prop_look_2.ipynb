```python
import pandas as pd
import date_utils as dt_mng
from data_manager import DataManager
```

```python
from IPython.display import display
pd.set_option('display.width', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
```

```python
dt = DataManager()
```

```python
analyzed_props = dt.get_analyzed_props()
filter_players = []
props = dt.filter_props(analyzed_props, filter_players, 100).reset_index(drop=True)
```

```python
prop_number = -1
```

```python
prop_number += 1
selected_prop = props.iloc[[prop_number]]
print(selected_prop)
player_id = dt.get_player_id(selected_prop['PLAYER'][prop_number])
games = dt.get_and_save_player_data(player_id)
display(games.head(25))
```

```python
wanted = [0, 1, 2, 3, 4, 9, 10, 11, 12, 16, 17, 21, 26, 27, 28, 33]
selected_props = props.iloc[wanted].reset_index(drop=True)
```

```python
selected_parlays = dt.select_optimal_parlays(selected_props, .33, .33, 3, 3)
print(selected_parlays)
```

```python
selected_parlay_dfs = []
for parlay in selected_parlays:
    parlay_rows = []
    for i in parlay['COMBO']:
        prop = selected_props.iloc[i]
        parlay_rows.append(prop)
    parlay_df = pd.concat(parlay_rows, axis=1, ignore_index=True).T
    parlay_df['PARLAY_EV'] = parlay['COMBINED_EV']
    parlay_df['PARLAY_PROB'] = parlay['COMBINED_PROB']
    parlay_df['PARLAY_HOUSE_PROB'] = parlay['COMBINED_HOUSE_PROB']
    parlay_df['PARLAY_EV'] = parlay['COMBINED_EV']
    parlay_df['PARLAY_ID'] = parlay['PARLAY_ID']
    parlay_df['TO_WIN'] = parlay['TO_WIN']
    selected_parlay_dfs.append(parlay_df)
selected_parlay_df = pd.concat(selected_parlay_dfs, axis=0, ignore_index=True)
selected_parlay_df['PROP_TAG'] = selected_parlay_df['PLAYER'] + "_" + selected_parlay_df['STAT'] + "_" + selected_parlay_df['TYPE']
print(selected_parlay_df)
```

```python
prop_distribution = dt.get_prop_distribution(selected_parlay_df)
player_distribution = dt.get_player_distribution(selected_parlay_df)
print(player_distribution)
print(prop_distribution)
print(len(selected_parlay_df['PARLAY_ID'].unique()))
```

```python
today_dir = dt_mng.get_or_create_directory_in_days()
selected_parlay_df.to_csv(f"{today_dir}/selected_parlay_df.csv")
```
