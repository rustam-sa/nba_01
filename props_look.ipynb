```python
dt = DataManager()
props = dt.get_analyzed_props()
```

```python
filter_props = ["Payton Pritchard", "Jayson Tatum", "T.J. McConnell", "Al Horford", "Obi Toppin", "Myles Turner"]
n_props = 100
props = dt.filter_props(props, filter_props, n_props)
```
